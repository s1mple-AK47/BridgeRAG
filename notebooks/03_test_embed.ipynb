```python
import requests
import json

response = requests.post(
    "http://localhost:8990/v1/embeddings",
    json={
        "model": "nomic-embed-text-v1",
        "input": ["Hello from vLLM!", "This is fast!"]
    }
)

if response.status_code == 200:
    data = response.json()
    embeddings = data['data'][0]['embedding']
    print(f"Successfully got embeddings. Shape of first one: {len(embeddings)}")
else:
    print("Error:", response.text)
```
